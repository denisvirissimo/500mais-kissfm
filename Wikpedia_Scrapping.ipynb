```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

df = pd.read_csv("./data/500+.csv")
df['Duracao'] = '0:00'

lista = df.drop_duplicates(subset='Musica_Wiki').Musica_Wiki.to_list()

for musica in lista:
		try:
			response = requests.get(url=musica)
			soup = BeautifulSoup(response.content, 'html.parser')
			duracao = soup.select('.duration')[0].get_text()
			print(musica, duracao)
			df.loc[df['Musica_Wiki'] == musica, 'Duracao'] = duracao
		except:
			df.loc[df['Musica_Wiki'] == musica, 'Duracao'] = 'erro'

df.to_csv('file.csv')
```
